```python
def check_pal(s):
    if s==s[::-1]:
        return 'pal'
    else:
        return 'not pal'
check_pal('mom')
```

```python
def check_pal():
    s=input()
    if s==s[::-1]:
        return 'pal'
    else:
        return 'not pal'
check_pal()
```

```python
def pat(n):
    for i in range(n):
        for j in range(2**i):
            print('a',end='')
        for j in range(i+1):
            print('b',end='')
        print()
pat(5)
```
