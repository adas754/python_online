```python
# check string is palindrom or not using function
# check number is even or odd using function
def check_palindrom(a):
    if a==a[::-1]:
        return 'palindrom'
    else:
        return 'not palindrom'
```

```python
check_palindrom('mama')
```

```python
# check the no is even or odd using the lambda
even_odd=lambda x:'even' if x%2==0 else 'odd'
even_odd(3)
```

```python
# try to find out maximum number in between two number
# the no is +ve , -ve or 0
max_of_two_numbers = lambda a,b : a if a>b else b
max_of_two_numbers(19,4)
```

```python
seq=[8,9,7,5,2,4]
# even 
list(filter(lambda x:x%2==0,seq))
```

```python
# try to find out 
# from this seq which one is greater than 2
seq=[6,5,1,9,0.5]
list(filter(lambda x:x>2,seq))
```

```python
prog=['python','php','java','c','c++']
# you have to find out the string which is started with p
```

```python
seq2 = ['python', 'php', 'java', 'c', 'c++']

list(filter(lambda x : x.startswith('p'), seq2))
```

```python
list(filter(lambda s:s.startswith('p'),prog))
```
